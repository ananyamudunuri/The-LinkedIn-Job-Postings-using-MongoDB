### Our objective is to create two final dataframes as follows. We will be taking subset from jobpostings to create job and jobpost datasets and eventually by merging data from other datasets will create final two.
       - 1. CompanyJobDetail : This dataframe will contain job details from from each company along with company details in embedded format
       - 2. JobPostFromCompany: This dataframe will contain details regarding job post from each company. This data is for job post details from linkedin platform and company employee count and follwer details.

###  1. CompanyJobDetail dataset creation

```python
job=job_postings_csv[['job_id', 'company_id', 'title', 'max_salary', 'min_salary',
       'pay_period', 'formatted_work_type', 'location', 
       'application_type', 'formatted_experience_level','sponsored', 'work_type']]
```

```python
job_post=job_postings_csv[['job_id', 'company_id', 'applies','views', 'job_posting_url', 'application_url','application_type']]
```

```python
### We will be joining dataframes - job and company
```

```python
CompanyJob_merge1=pd.merge(job, company_csv,
                   on='company_id',  
                   how='left')
```

### We will process job_skills_csv : We will list all the  job skills for each job id and store them in a column

```python
#Creating array of skills in skill_abr column
job_skills=job_skills_csv.groupby('job_id')['skill_abr'].apply(list).reset_index()
```

```python
job_skills.head()
```

```python
CompanyJob_merge2=pd.merge(CompanyJob_merge1, job_skills,
                   on='job_id',  
                   how='left')
```

```python
CompanyJob_merge2.head()
```

### We will process company_industries_csv : We will list all the  industries for each company id and store them in a column

```python
company_industries_csv.head()
```

```python
company_industries=company_industries_csv.groupby('company_id')['industry'].apply(list).reset_index()
```

```python
company_industries.head()
```

## Let us merge company_industries with main dataframe

```python
CompanyJob_merge3=pd.merge(CompanyJob_merge2, company_industries,
                   on='company_id',  
                   how='left')
```

```python
CompanyJob_merge3.head()
```

### We will process company_specialities_csv : We will list of speciality for each company id and store them in a column

```python
company_specialities_csv.head()
```

```python
company_specialities=company_specialities_csv.groupby('company_id')['speciality'].apply(list).reset_index()
```

```python
company_specialities.head()
```

## Let us merge company_specialities with main dataframe

```python
CompanyJob_merge4=pd.merge(CompanyJob_merge3, company_specialities,
                   on='company_id',  
                   how='left')
```

```python
CompanyJob_merge4.head()
```

### Renaming title column as job_title, name column as company_name, location as job_location

```python
CompanyJob_merge4.rename(columns = {'title':'job_title','name':'company_name','location':'job_location'}, inplace = True)
```

```python
# Checking null values in each columns
CompanyJob_merge4.isnull().sum()
```

```python
CompanyJob_merge4.info()
```

### We will take company data in another dataset

```python
CompanyJob_merge4['company'] = CompanyJob_merge4[['company_name','description','company_size','state','country','city','zip_code','url','industry','speciality']].to_dict('records')
```

```python
CompanyJobDetail=CompanyJob_merge4[['job_title','company','max_salary','min_salary','pay_period','formatted_work_type','job_location','application_type','formatted_experience_level','sponsored','work_type','skill_abr']]
```

```python
CompanyJobDetail.head()
```

### converting into JSON format

```python
CompanyJobDetail.to_json('~\Documents\Data_225_Lab2\LinkedinJobAnalysis_CompanyJobDetail.json', orient='records', lines=True)
```

### 2. JobPostFromCompany dataset creation

```python
benefits_csv.columns
```

```python
### We will process benefits_csv : We will list of benefit type for each job id and store them in a column
```

```python
benefits=benefits_csv.groupby('job_id')['type'].apply(list).reset_index()
```

### We will merge job_post and benefits 

```python
Jobpost_merge1=pd.merge(job_post, benefits,
                   on='job_id',  
                   how='left')
```

```python
Jobpost_merge1.head()
```

### Let us merge company name from company dataset with Jobpost_merge1 dataset

```python
Jobpost_merge2=pd.merge(Jobpost_merge1,company_csv[['company_id','name']],on='company_id',how='inner')
```

```python
Jobpost_merge2.head()
```

### We will merge employee_counts_csv with  Jobpost_merge2

```python
Jobpost_merge3=pd.merge(Jobpost_merge2,employee_counts_csv[['company_id','employee_count','follower_count']],on='company_id',how='left')
```

```python
Jobpost_merge3.head()
```

### Let us merge job title from job_title dataset with Jobpost_merge3 dataset

```python
Jobpost_merge4=pd.merge(Jobpost_merge3,job_postings_csv[['job_id','title']],on='job_id',how='inner')
```

### Renaming type column as benefit_type, name column as company_name

```python
Jobpost_merge4.rename(columns = {'type':'benefit_type','title':'job_title','name':'company_name'}, inplace = True)
```

### We will take company data in another dataset

```python
Jobpost_merge4['company_details'] = Jobpost_merge4[['company_name','employee_count','follower_count']].to_dict('records')
```

```python
Jobpost_merge4.head()
```

```python
JobPostFromCompany=Jobpost_merge4[['job_title','company_details','applies','views','job_posting_url','application_url','application_type','benefit_type']]
```

```python
JobPostFromCompany.head()
```

### converting into JSON format

```python
JobPostFromCompany.to_json('~\Documents\Data_225_Lab2\LinkedinJobAnalysis_JobPostFromCompany.json', orient='records', lines=True)
```
